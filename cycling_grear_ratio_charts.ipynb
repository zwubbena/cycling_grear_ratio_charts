```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Defining chainrings and sprockets
chainrings = [52, 36]
sprockets = [11-12-13-14-15-16-17-19-21-24-27-30]

# Calculating gear ratios
gear_ratios = [[chainring/sprocket for sprocket in sprockets] for chainring in chainrings]

# Creating a heatmap for the gear ratio chart
plt.figure(figsize=(12, 5))
sns.heatmap(gear_ratios, annot=True, fmt=".2f", cmap="coolwarm", xticklabels=sprockets, yticklabels=chainrings)
plt.title("Gear Ratio Chart")
plt.xlabel("Rear Cassette Sprockets")
plt.ylabel("Front Chainrings")
plt.show()

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Defining chainrings and sprockets
chainrings = [44]
sprockets = [11, 12, 13, 14, 15, 17, 19, 21, 24, 27, 30, 34]
rpm = 90

# Calculating gear ratios
gear_ratios = [[chainring/sprocket for sprocket in sprockets] for chainring in chainrings]

# Creating a heatmap for the gear ratio chart
plt.figure(figsize=(12, 5))
sns.heatmap(gear_ratios, annot=True, fmt=".2f", cmap="coolwarm", xticklabels=sprockets, yticklabels=chainrings)
plt.title("Gear Ratio Chart")
plt.xlabel("Rear Cassette Sprockets")
plt.ylabel("Front Chainrings")
plt.show()



import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Defining chainrings and sprockets
chainrings = [50, 34, 44, 46, 48]
sprockets = [11, 12, 13, 14, 15, 17, 19, 21, 24, 27, 30, 34]

# Calculating gear ratios
gear_ratios = [[chainring/sprocket for sprocket in sprockets] for chainring in chainrings]

# Creating a heatmap for the gear ratio chart
plt.figure(figsize=(12, 5))
sns.heatmap(gear_ratios, annot=True, fmt=".2f", cmap="coolwarm", xticklabels=sprockets, yticklabels=chainrings)
plt.title("Gear Ratio Chart")
plt.xlabel("Rear Cassette Sprockets")
plt.ylabel("Front Chainrings")
plt.show()
```
